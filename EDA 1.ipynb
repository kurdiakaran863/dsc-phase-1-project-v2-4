```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('final.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['genre3'].isnull().sum()
```

```python
df = df[df['release_date'] > 2009]
```

```python
df.describe()
```

```python
df['profit'] = df['worldwide_gross'] - df['production_budget']
df['profit']
```

```python
df.describe()
```

```python
df.drop('genre3', axis=1, inplace = True)
```

```python
df['genre2'].isnull().sum()
```

```python
#see count of rows where percentage is less then 0
df = df.dropna(subset = ['genre2'])
```

```python
df
```

```python
df.describe()
```

```python
avg_profit = df.groupby("genre2")["profit"].mean().sort_values(ascending= False)
avg_prof_year = df.groupby("release_date")["profit"].mean().sort_values(ascending= True)
avg_budget = df.groupby("genre2")["production_budget"].mean().sort_values(ascending= False)
#avg_ratio = df.groupby("genre1")["ratio"].mean().sort_values(ascending= False)
```

```python
#profit across generes
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(45, 30))
plt.rcdefaults()
# sets up bar plot
#y = df['profit']
#x = df['genre2']
plt.rcParams.update({'font.size':45})
ax.bar(avg_profit.keys(), avg_profit, color = "green")
ax.set_xlabel('Movie Genre')
ax.set_ylabel("Average Profit")
ax.set_title('Average Movie Profit per Genre from 2010 to 2019')
ax.set_xticklabels(avg_profit.keys(), rotation=90)
#ax.get_yaxis().set_major_formatter(
    #ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
#plt.rcParams.update({'font.size': 10})
plt.grid()
plt.tight_layout()
plt.show();
```

```python
#genre vs budget

fig, ax = plt.subplots(figsize=(45, 30))
#plt.rcdefaults()
# sets up bar plot
plt.rcParams.update({'font.size':45})
ax.bar(avg_budget.keys(), avg_budget, color = "green")
ax.set_xlabel('Movie Genre')
ax.set_ylabel("Average Budget")
ax.set_title('Average Movie Budget per Genre from 2010 to 2019')
ax.set_xticklabels(avg_budget.keys(), rotation=90)
#ax.get_yaxis().set_major_formatter(
    #ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
#plt.rcParams.update({'font.size': 10})
plt.grid()
plt.tight_layout()
plt.show();
```

```python
df['profit_to_budget'] = (df['profit'] / df['production_budget'])
df
```

```python
profit_to_budget = df.groupby("genre2")["profit_to_budget"].mean().sort_values(ascending= False)
profit_to_budget
```

```python
# which genere has the best profit to budget performance
fig, ax = plt.subplots(figsize=(45, 30))
#plt.rcdefaults()
# sets up bar plot
plt.rcParams.update({'font.size':45})
ax.bar(profit_to_budget.keys(), profit_to_budget, color = "blue")
ax.set_xlabel('Movie Genre')
ax.set_ylabel("profit to budget ratio")
ax.set_title('profit to budget performance per Genre from 2010 to 2019')
ax.set_xticklabels(profit_to_budget.keys(), rotation=90)
#ax.get_yaxis().set_major_formatter(
    #ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
#plt.rcParams.update({'font.size': 10})
plt.grid()
plt.tight_layout()
plt.show();
```
