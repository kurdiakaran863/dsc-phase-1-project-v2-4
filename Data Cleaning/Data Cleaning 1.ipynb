```python
import pandas as pd
import sqlite3 
```

```python
conn = sqlite3.connect("im.db")
cur = conn.cursor() 
```

```python
#viewing content of data in a tabular form(dataframe)
movie_basics = pd.read_sql(
"""
SELECT *
FROM movie_basics
"""
, conn)
movie_basics.dtypes
```

```python
movie_basics['movie_id'] = movie_basics['movie_id'].astype(str)
movie_basics['primary_title'] = movie_basics['primary_title'].astype(str)
movie_basics['genres'] =  movie_basics['genres'].tolist()
movie_basics.info()
```
