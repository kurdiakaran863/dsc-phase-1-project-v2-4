```python
import pandas as pd
```

```python
the_number = pd.read_csv("tn.movie_budgets.csv")
the_number.head()
```

```python
the_number.info()
```

```python
#converting the object type into int for numerical values
the_number['production_budget'] = the_number['production_budget'].str.replace(',', '').str.replace('$', '').astype(int)
the_number['domestic_gross'] = the_number['domestic_gross'].str.replace(',', '').str.replace('$', '').astype(int)
the_number['worldwide_gross'] = the_number['worldwide_gross'].str.replace(',', '').str.replace('$', '').astype(int)
the_number.head()
```

```python
#looking at the data types to verify
the_number.dtypes
```

```python
#checking if there are any null values
the_number.isnull().sum()
```

```python
#removing rows where column value is 0
the_number = the_number.loc[(the_number['domestic_gross'] > 0) & (the_number['worldwide_gross'] > 0)]
the_number
```

```python
the_number['movie'] = the_number['movie'] .str.replace('[^\w\s]','')
the_number.head()
```

```python
movie_table = pd.read_csv("new_movie_table.csv")
movie_table.shape
```

```python
movie_table['primary_title'] = movie_table['primary_title'] .str.replace('[^\w\s]','')
```

```python
movie_table['primary_title']
```

```python
the_number = the_number.rename(columns={'movie': 'primary_title'})
```

```python
the_number
```

```python
new_movie_table = pd.merge(the_number,movie_table, on='primary_title', how="inner")
```
