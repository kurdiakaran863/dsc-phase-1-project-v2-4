```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('final.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['genre3'].isnull().sum()
```

```python
df = df[df['release_date'] > 2009]
```

```python
df.describe()
```

```python
df['profit'] = df['worldwide_gross'] - df['production_budget']
df['profit']
```

```python
df['production_budget'] = df['production_budget'] / 1000000
df['domestic_gross'] = df['domestic_gross'] / 1000000
df['worldwide_gross'] = df['worldwide_gross'] / 1000000
df['profit'] = df['profit'] / 1000000
df.head()
```

```python
df.describe()
```

```python
df.drop('genre3', axis=1, inplace = True)
```

```python
df['genre2'].isnull().sum()
```

```python
#see count of rows where percentage is less then 0
df = df.dropna(subset = ['genre2'])
```

```python
df
```

```python
df.describe()
```

```python
avg_profit = df.groupby("genre2")["profit"].mean().sort_values(ascending= False)
```

```python

import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(55, 30))
plt.rcdefaults()
plt.rcParams.update({'font.size':55})
ax.bar(avg_profit.keys(), avg_profit, color = "green")
ax.set_xlabel("Movie Genre")
ax.set_ylabel('Profit in mil')
ax.set_title('Movie Profit per Genre from 2010 to 2019')
ax.set_xticklabels(avg_profit.keys(), rotation=30)
plt.grid()
plt.tight_layout()
plt.show();
```

```python
top5_df = df.loc[(df["genre2"] == "Adventure") | (df["genre2"] == "Animation") | (df["genre2"] == "Fantasy") | (df["genre2"] == "Family") |(df["genre2"] == "Sci-Fi")]
```

```python
top5_df
```

```python
top5_df.drop("genre1", axis=1, inplace=True)
```

```python
top5_df.rename(columns = {'genre2':'genre'}, inplace = True)
```

```python
top5_df
```

```python
top5_df['genre'].dtypes

```

```python
import seaborn as sns
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(70,30))
sns.boxplot(data=top5_df, x='genre', y='profit', showmeans=True)
plt.rcParams.update({'font.size':75})
ax.set_xlabel('Movie Genre')
ax.set_ylabel("Profit in mil")
ax.set_title(' generes based on the profit distribution ')
ax.get_yaxis().set_major_formatter(
    ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.tight_layout()
plt.show();
```

```python
import seaborn as sns
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(15, 9))
sns.scatterplot(data=top5_df, x='production_budget', y='profit', hue ='genre', palette='Blues_r',)
plt.rcParams.update({'font.size':25})
ax.set_xlabel('Budget in mil')
ax.set_ylabel("Profit in mil")
ax.set_title('Distribution of Budget vs Profit gained')
plt.grid()
plt.tight_layout()
plt.show();
```
